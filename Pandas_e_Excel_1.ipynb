## Biblioteca XlsxWriter para manipular arquivos XLSX em Python Pandas

Com a biblioteca XlsxWriter é possível escrever em arquivos xlsx, do Excel e equivalentes.

### Instalação
A biblioteca está disponível no repositório padrão do Anaconda:  

    conda install xlsxwriter

Também é possível pelo PIP:

    pip install XlsxWriter
    
Outras opções na página: https://xlsxwriter.readthedocs.io/getting_started.html#getting-started

### Exemplo simples

```python
import xlsxwriter
```

```python
workbook = xlsxwriter.Workbook('hello.xlsx')
worksheet = workbook.add_worksheet()

worksheet.write('A1', 'Hello world')

workbook.close()
```

### Salvando um dataframe do pandas em um arquivo xlsx
Aqui estou usando o dataset `iris` do *scikit-learn* previamente salvo como csv.

```python
import pandas as pd
iris_df = pd.read_csv('iris.csv')
iris_df.head()
```

Criando o objeto `writer` para receber os dados a serem salvos no arquivo xlsx.  
É preciso informar o *engine* XlsxWriter.  
Após, converter o dataframe para um objeto XlsxWriter.  

```python
writer = pd.ExcelWriter('iris_df.xlsx', engine='xlsxwriter')
iris_df.to_excel(writer, sheet_name='Sheet1')
```

O objeto `writer` está na memória, é preciso salvar no arquivo xlsx:

```python
writer.save()
```
